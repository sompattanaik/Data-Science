```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv('Universities.csv')
data.head(5)
```

```python
data.shape
```

```python
# Considering only numerical data
uni = data.iloc[:,1:]
data_u = uni.values
data_u
```

```python
# standardize the data

data_std = scale(data_u)
data_std
```

```python
pca = PCA(n_components=6)
pca_vals = pca.fit_transform(data_std)
pca_vals
```

```python
pca_vals.shape
```

So after applying pca our new normalized data have 25 rows and 6 features.

```python
#weights

pca.components_
```

```python
# variances that each principal components explains:

var = pca.explained_variance_ratio_
var
```

```python
# cumulative sum of variance :

var1 = np.cumsum(np.round(var,4)*100)
var1
```

here we can tell that 1st 2 components are giving me 90% of explained variance.

```python
plt.plot(var1,c = 'g')
```

```python
final_df = pd.concat([pd.DataFrame(pca_vals[:,:2],columns = ['pc1','pc2']),data[['Univ']]],axis = 1)
final_df
```

```python
sns.scatterplot(data = final_df,x = 'pc1',y = 'pc2', hue = 'Univ')
```

```python
p = sns.scatterplot(data = final_df,x = 'pc1',y = 'pc2')
for i in range (0,final_df.shape[0]):
  p.text(final_df.pc1[i],final_df.pc2[i],final_df.Univ[i],horizontalalignment = 'left')
```
